```python
import re
import pandas as pd
import os

def check_parquet_file(file_path):
    """
    Checks if a parquet file name fits a specific format (student_id.parquet) and its contents meet certain criteria.

    Args:
        file_path: The path to the parquet file.

    Returns:
        True if the file name and contents meet the criteria, False otherwise.
        Prints informative messages about the checks.
    """

    # Check file name format
    filename = os.path.basename(file_path)
    pattern = r"^\d+\.parquet$"  # Matches any number of digits (student id) followed by .parquet
    if not re.match(pattern, filename):
        print(f"Error: Invalid file name format. Expected format: digits.parquet, got {filename}")
        return False

    try:
        # Read parquet file into pandas DataFrame
        df = pd.read_parquet(file_path)

        # Check DataFrame shape
        if df.shape != (222, 14):
            print(f"Error: Invalid DataFrame shape. Expected shape: (222, 14), got {df.shape}")
            return False

        # Check for -100 values in Column 'G'
        if (df['G'] == -100).any():
            print("Warning: Column 'G' contains -100 values. Predictive modeling needed.")
            return False # Return true because the goal is to replace -100s, and they exist

        print("File name and content checks passed. GOOD !")
        return True
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return False
    except Exception as e:
        print(f"An error occurred: {e}")
        return False


# Example usage: replace with actual file path
file_path = "/content/drive/MyDrive/Courses/CMPE 301/Assignment/modified.parquet" # Example path

check_parquet_file(file_path)
```
